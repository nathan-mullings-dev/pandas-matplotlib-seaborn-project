## loc and iloc

```python
# return name, gender and class column student with their id from 10 to 20 using loc
df.loc[9:19,['id','name','gender','class']]
```

```python
# return name, gender and class column student with their id from 10 to 20 using iloc
df.iloc[10:20,[0,1,4,2]]
```

```python
# use loc operator to return mark of the student at index 21
df.loc[21]['mark']
```

```python
# use iloc operator to return mark of the student at index 21
df.iloc[21]['mark']
```

```python
# use loc method to return average mark of the first 10 students .
df.loc[0:10, 'mark'].mean()
```

```python
# use iloc method to return average mark of the first 10 students .
df.iloc[0:10, df.columns.get_loc('mark')].mean()
```

## filtering data

```python
# use slicing method to filter female students
df[df['gender'] == 'female']
```

```python
# use loc operator to filter female students
df.loc[df['gender'] == 'female']
```

```python
# use slicing method to return students who has their mark>=85
df[df['mark'] > 85]
```

```python
# use slicing method to return  only name, class and mark of the students who has their mark>=85
df[df['mark'] > 85][['name','class','mark']]
```

```python
# use loc method to return students who has their mark>=85
df.loc[df['mark'] > 85]
```

```python
# # use loc operator to return  only name, class and mark of the students who has their mark>=85
df.loc[df['mark'] > 85, ['name','class','mark']]
```

```python
# use slicing method to return  average mark of the students who has their mark>=85
df[df['mark'] > 85]['mark'].mean()
```

```python
# use loc operator to return average mark of the students who has their mark>=85
df.loc[df['mark'] > 85, 'mark'].mean()
```

```python
# use slicing method to return female students at class four
df[(df['gender'] == 'female') & (df['class'] == 'Four')]
```

```python
# use loc operator to return female students at class four
df.loc[(df['gender'] == 'female') & (df['class'] == 'Four')]
```

```python
# How would you find out how many distinct marks are recorded and their frequency?
df['mark'].value_counts()
```

```python
# How would you sort the dataset first by class in descending order and then by mark in ascending order within each class?
df.sort_values(by=['class', 'mark'], ascending=[False, True])
```
